```python
import pandas as pd
import numpy as np
import matplotlib as mp
import seaborn as sb
import math as mt 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('electronics.csv')
print(data)
```

```python
data.describe()
```

```python
#1. In which year did the best sales occur?

print(f"The year with the best sales is {data.groupby('year')['rating'].sum().idxmax()} with {data.groupby('year')['rating'].sum().max()} sales.")
print(f"The year with the best sales is {data.groupby('year')['rating'].count().idxmax()} with {data.groupby('year')['rating'].count().max()} sales.")

data.groupby('year')['rating'].count().plot(kind='bar')
```

```python
'''#2. In which month did the best sales occur?'''
data.groupby('month')['rating'].count().plot(kind='bar')
```

```python
# 3.which brands sold most.

data.groupby('brand')['rating'].count().sort_values(ascending=False).head(15).plot(kind='bar')
```

```python
# 4. which category sold most.
data.groupby('category')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# 5. which brand sold the least.
data.groupby('brand')['rating'].count().sort_values(ascending=True).head(10).plot(kind='bar')
```

```python
#6. which category sold the least.
data.groupby('category')['rating'].count().sort_values(ascending=True).head(10).plot(kind='bar')
```
